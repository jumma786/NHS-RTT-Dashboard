```python
import pandas as pd
import numpy as np

file_path = r"C:\Users\jumma\Downloads\NHS-RTT-Dashboard\dashboard\NHS_RTT_Dashboard.xlsx"

df = pd.read_excel(file_path, sheet_name="Data")

# Convert date column
df['Period'] = pd.to_datetime(df['Period'])

# Convert wait columns to numeric
df['Median_Wait_Wks'] = pd.to_numeric(df['Median_Wait_Wks'], errors='coerce')
df['P92_Wait_Wks'] = pd.to_numeric(df['P92_Wait_Wks'], errors='coerce')

# Check data types
print(df.dtypes)

# Missing values after conversion
print("\nMissing values:")
print(df[['Median_Wait_Wks', 'P92_Wait_Wks']].isnull().sum())
```

```python
# National KPI Summary

print("Total Waiting List:", df['Incomplete_Pathways'].sum())
print("Total Treated Within 18 Weeks:", df['Within_18wks'].sum())
print("Average % Within 18 Weeks:", round(df['Pct_Within_18wks'].mean()*100, 2))
print("Average Median Wait:", round(df['Median_Wait_Wks'].mean(), 2))
print("Total 52+ Week Waiters:", df['Waiters_52plus'].sum())
print("Total 65+ Week Waiters:", df['Waiters_65plus'].sum())
print("Total 78+ Week Waiters:", df['Waiters_78plus'].sum())
```

```python
icb_waiting = (
    df.groupby('ICB_Name')['Incomplete_Pathways']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

print(icb_waiting)
```

```python
specialty_waits = (
    df.groupby('Treatment_Function')['Median_Wait_Wks']
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

print(specialty_waits)
```

```python
monthly_trend = (
    df.groupby('Period')['Incomplete_Pathways']
      .sum()
      .reset_index()
)

monthly_trend.head()
```

```python
df['Incomplete_Pathways'].sum()
```

```python
latest_period = df['Period'].max()

latest_df = df[df['Period'] == latest_period]

print("Latest Period:", latest_period)

print("Current Waiting List:",
      latest_df['Incomplete_Pathways'].sum())

print("Current 52+ Week Waiters:",
      latest_df['Waiters_52plus'].sum())

print("Current 65+ Week Waiters:",
      latest_df['Waiters_65plus'].sum())

print("Current 78+ Week Waiters:",
      latest_df['Waiters_78plus'].sum())

print("Current RTT Performance:",
      round(latest_df['Pct_Within_18wks'].mean()*100,2), "%")
```

```python
monthly_trend = (
    df.groupby('Period')
      .agg({
          'Incomplete_Pathways':'sum',
          'Waiters_52plus':'sum'
      })
      .reset_index()
)

monthly_trend.tail()
```

```python
start = monthly_trend.iloc[0]['Incomplete_Pathways']
end = monthly_trend.iloc[-1]['Incomplete_Pathways']

growth = ((end - start) / start) * 100

print(f"Waiting List Growth: {growth:.2f}%")
```

```python
latest_icb = (
    latest_df.groupby('ICB_Name')
    .agg({
        'Pct_Within_18wks':'mean',
        'Incomplete_Pathways':'sum'
    })
    .sort_values('Pct_Within_18wks')
)

latest_icb.head(10)
```

```python
latest_specialties = (
    latest_df.groupby('Treatment_Function')
    .agg({
        'Median_Wait_Wks':'mean',
        'Incomplete_Pathways':'sum'
    })
    .sort_values('Median_Wait_Wks', ascending=False)
)

latest_specialties.head(10)
```

```python
import matplotlib.pyplot as plt

monthly_trend.plot(
    x='Period',
    y='Incomplete_Pathways',
    figsize=(12,6)
)

plt.title('NHS Waiting List Trend')
plt.ylabel('Incomplete Pathways')
plt.grid(True)
plt.show()
```

```python
long_waits = df.groupby('Period')[['Waiters_52plus',
                                   'Waiters_65plus',
                                   'Waiters_78plus']].sum()

long_waits.plot(figsize=(12,6))
plt.title('Long Waiters Trend')
plt.ylabel('Patients')
plt.grid(True)
plt.show()
```

```python
rtt_trend = df.groupby('Period')['Pct_Within_18wks'].mean()*100

plt.figure(figsize=(12,6))
plt.plot(rtt_trend.index, rtt_trend.values)
plt.title('RTT Performance Trend')
plt.ylabel('% Within 18 Weeks')
plt.grid(True)
plt.show()
```

```python
top_icbs = latest_df.groupby('ICB_Name')['Incomplete_Pathways']\
                    .sum()\
                    .sort_values(ascending=False)\
                    .head(10)

top_icbs.sort_values().plot(
    kind='barh',
    figsize=(10,6)
)

plt.title('Top 10 ICBs by Waiting List')
plt.show()
```

```python
worst_icbs = latest_df.groupby('ICB_Name')['Pct_Within_18wks']\
                      .mean()\
                      .sort_values()\
                      .head(10)

(worst_icbs*100).plot(
    kind='barh',
    figsize=(10,6)
)

plt.title('Worst RTT Performance by ICB')
plt.xlabel('% Within 18 Weeks')
plt.show()
```

```python
specialty_backlog = latest_df.groupby(
    'Treatment_Function'
)['Incomplete_Pathways'].sum()

specialty_backlog.sort_values().plot(
    kind='barh',
    figsize=(12,8)
)

plt.title('Waiting List by Specialty')
plt.show()
```

```python
specialty_waits = latest_df.groupby(
    'Treatment_Function'
)['Median_Wait_Wks'].mean()

specialty_waits.sort_values().plot(
    kind='barh',
    figsize=(12,8)
)

plt.title('Median Wait Time by Specialty')
plt.xlabel('Weeks')
plt.show()
```

```python
wait52 = latest_df.groupby(
    'Treatment_Function'
)['Waiters_52plus'].sum()

wait52.sort_values().tail(10).plot(
    kind='barh',
    figsize=(10,6)
)

plt.title('52+ Week Waiters by Specialty')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(
    latest_df['Pct_Within_18wks']*100,
    bins=20
)

plt.title('Distribution of RTT Performance')
plt.xlabel('% Within 18 Weeks')
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    latest_df['Median_Wait_Wks'],
    latest_df['Pct_Within_18wks']*100
)

plt.xlabel('Median Wait (Weeks)')
plt.ylabel('RTT Performance (%)')
plt.title('Wait Time vs RTT Performance')

plt.show()
```

```python
top_specialties = latest_df.groupby(
    'Treatment_Function'
)['Incomplete_Pathways'].sum()\
 .nlargest(8)

plt.figure(figsize=(8,8))
plt.pie(
    top_specialties,
    labels=top_specialties.index,
    autopct='%1.1f%%'
)

plt.title('Waiting List Share by Specialty')
plt.show()
```

```python
monthly_trend['Monthly_Change'] = \
monthly_trend['Incomplete_Pathways'].diff()

plt.figure(figsize=(12,6))
plt.bar(
    monthly_trend['Period'],
    monthly_trend['Monthly_Change']
)

plt.title('Monthly Waiting List Change')
plt.show()
```

```python
import seaborn as sns

pivot = latest_df.pivot_table(
    values='Pct_Within_18wks',
    index='ICB_Name',
    columns='Treatment_Function',
    aggfunc='mean'
)

plt.figure(figsize=(16,10))
sns.heatmap(pivot, cmap='RdYlGn')
plt.title('RTT Performance by ICB and Specialty')
plt.show()
```
